```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
data_top = np.load("data/early.npy")
data_bottom = np.load("data/late.npy")
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

im1 = ax1.imshow(
    data_top, 
    cmap='plasma', 
    aspect='auto', 
    extent=[0, 10000, 50, 0], 
    interpolation='nearest'
)
ax1.set_title('Average Q-Values over Episodes (Early Training)')
ax1.set_xlabel('Episode')
ax1.set_ylabel('Steps from goal state')

divider1 = make_axes_locatable(ax1)
cax1 = divider1.append_axes("right", size="5%", pad=0.15)
cbar1 = fig.colorbar(im1, cax=cax1)

im2 = ax2.imshow(
    data_bottom, 
    cmap='plasma', 
    aspect='auto', 
    extent=[300000, 400000, 50, 0], 
    interpolation='nearest'
)
ax2.set_title('Average Q-Values over Episodes (Late Training)')
ax2.set_xlabel('Episode')
ax2.set_ylabel('Steps from goal state')

divider2 = make_axes_locatable(ax2)
cax2 = divider2.append_axes("right", size="5%", pad=0.15)
cbar2 = fig.colorbar(im2, cax=cax2)

plt.tight_layout()
plt.show()
```

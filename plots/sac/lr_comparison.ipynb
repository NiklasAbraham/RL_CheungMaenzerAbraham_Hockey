```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
series = pd.read_pickle('data/lr_series.pkl')
```

```python
fig, ax = plt.subplots(figsize=(12, 6))

lrs = [0.0003, 0.001, 0.005, 0.01]
colors = ['#4C78A8', '#F58518', '#54A24B', '#E45756']

for lr, color in zip(lrs, colors):
    s = series[lr]
    ax.plot(s["step"], s["winrate"], marker='o', linewidth=2.2, color=color, label=f"{lr}")

ax.set_title('Winrate over Steps')
ax.set_xlabel('Step')
ax.set_ylabel('Winrate')
ax.set_ylim(0.0, 1.0)
ax.legend(title='Learning rate')
ax.grid(True)

plt.tight_layout()
plt.show()
```

```python
from rl_hockey.common.archive import Archive
import os
```

```python
archive = Archive("archive")

base_path = "results/sac_runs/2026-02-25_01-27-48"

checkpoint_files = [f for f in os.listdir(os.path.join(base_path, "models")) if f.endswith(".pt")]

for i, checkpoint in enumerate(checkpoint_files):
    if i == len(checkpoint_files) - 1:
        checkpoint_path = os.path.join(base_path, "models", checkpoint)
        config_path = os.path.join(base_path, "config.json")
        archive.add_agent(checkpoint_path, config_path, checkpoint[:-3])
```

```python
# replace \\ with /
with open(os.path.join("archive", "registry.json"), "r") as f:
    data = f.read()
data = data.replace("\\\\", "/")
with open(os.path.join("archive", "registry.json"), "w") as f:
    f.write(data)

for dir in os.listdir(os.path.join("archive", "agents")):
    file_path = os.path.join("archive", "agents", dir, "metadata.json")
    with open(file_path, "r") as f:
        data = f.read()
    data = data.replace("\\\\", "/")
    with open(file_path, "w") as f:
        f.write(data)
```
